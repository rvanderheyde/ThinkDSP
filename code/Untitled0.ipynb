```python
import numpy
import thinkdsp
import thinkplot
import math

signal = thinkdsp.TriangleSignal()
wave = signal.make_wave(duration=2)
ys = wave.ys
dfts = numpy.fft.fft(ys)
freqs = numpy.fft.fftfreq(ys.shape[-1])
diffs = dfts*2*math.pi*1j*freqs
new_ys = numpy.fft.ifft(diffs)

thinkplot.plot(new_ys)
```
