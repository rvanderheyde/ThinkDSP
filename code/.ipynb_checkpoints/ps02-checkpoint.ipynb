1) Run and listen to the examples in chap03.ipynb.  In the leakage example, try replacing the Hamming window with one of the other windows provided by NumPy, and see what effect they have on leakage. See http://docs.scipy.org/doc/numpy/reference/routines.window.html

```python
import thinkdsp
import thinkplot
import numpy
from matplotlib import pyplot
%matplotlib inline

signal = thinkdsp.SinSignal(freq=440)
duration = signal.period * 30.3
wave = signal.make_wave(duration)

wave.make_spectrum().plot(high=50)
bartlett = numpy.bartlett(len(wave.ys))
wave.window(bartlett)
spectrum = wave.make_spectrum()
pyplot.figure()
spectrum.plot(high=50)
```

Originally I used a duration of $30.5*period$ but the window's effect was to hard to see, which is why I moved to $30.3*period$ for the duration. Clearly the bartlett window did a better job reducing leakage on the lower frequencies than on the higher ones.

2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.

Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.

```python
import math
PI2 = math.pi *2
class SawtoothChirp(thinkdsp.Chirp):
    def __init__(self, start=220, end=880, amp=1.0):
        """Initializes a linear chirp.

        start: float frequency in Hz
        end: float frequency in Hz
        amp: float amplitude, 1.0 is nominal max
        """
        self.start = start
        self.end = end
        self.amp = amp
        
    def evaluate(self, ts):
        freqs = numpy.linspace(self.start, self.end, len(ts)-1)
        return self._evaluate(ts, freqs)
    
    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dps = PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        frac,_ = numpy.modf(phases)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac),self.amp)
        return ys

signal = SawtoothChirp()
wave = signal.make_wave(duration=5, framerate=11025)
spectrogram = wave.make_spectrogram(seg_length = 512)
spectrogram.plot()
```

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

If the length of the tube changes by $l$ per second, the resulting pitch changes at $\frac{1}{l}$ per second. Thus, the frequency varies quickly at first then slower and slower, and is more like a negative exponential chirp than a linear one.

```python
def glissando(start=262.0, end=349.0):
#     dp = 1/dl
    if end>start:
        dl = end/start
    else:
        dl = start/end
#     t =.266
    freqs = []
    freqs[0] = start
    for j in range(0,dl*100+1):
        i = j/100
        freqs.append(start/i)
    for j in range(dl*100+1,0,-1):
        i= j/100
        freqs.append(start/i)
    return freqs
        
```

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.

```python
import thinkdsp
import numpy
import matplotlib.pyplot as pyplot
sound = thinkdsp.read_wave('NorthSea.wav')
specStore = []
seg = sound.segment(0,60)
for i in range(10,70,10):
    spec = seg.segment(i-10,i).make_spectrum()
    ps = spec.power
    pers = [x/len(spec) for x in ps]
    specStore.append(pers)
total = []  
dic = {}
for lis in specStore:
    for i, val in enumerate(lis):
        if i not in dic:
            dic[i] = val
        else:
            dic[i] += val
for i in range(len(dic)):
    total.append(dic[i]/len(spec))

spec = seg.make_spectrum()
fs = numpy.array(spec.fs[0:661502])
cs = numpy.array(total)
pyplot.plot(cs,fs)
pyplot.yscale('log')
pyplot.xscale('log')
pyplot.show()
```
