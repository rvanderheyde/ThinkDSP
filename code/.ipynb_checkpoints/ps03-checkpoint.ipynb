1). Compute the power spectrum of BitCoin prices. Does it resemble white, pink, or Brownian noise?

```python
import pandas
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
wave = thinkdsp.Wave(ys, framerate=1)
spectrum = wave.make_spectrum()
spectrum.plot_power(low=1, linewidth=1, alpha =.5)
thinkplot.show(xlabel = 'frequency', ylabel = 'power density', xscale = 'log',yscale = 'log')
slope = spectrum.estimate_slope()[0]
print slope
```

scipy.stats.linregress on the log of power and log of frequency returns -1.8 which is close to -2, the slope that is expected if the data resembled brownian noise. Because the price depends on the previous price, this makes some sense as brownian noise has the characteristic that the next value is the previous value plus a random step.

2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    """Coefficient of correlation.

    ddof=0 indicates that we should normalize by N, not N-1.

    xs: sequence
    ys: sequence

    returns: float
    """
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    """Computes serial correlation with given lag.

    wave: Wave
    lag: integer, how much to shift the wave

    returns: float correlation coefficient
    """
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags, corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
```

So it looks like the autocorrelation decays exponentially at first then reaches a sort of periodic behavior but with each peak still decaying.

4) Test the algorithmic complexity of analyze1, analyze2 and scipy.fftpack.dct.

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
import dct
ns = 2 ** numpy.arange(4, 12)
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze1(ys, freqs, ts)
    results.append(result)
bests = [result.best for result in results]
bests
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
t
```

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:
Break a long signal into segments.
Compute the DCT of each segment.
Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.
Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible?

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')

def breakWave(wave, num=20, duration=20):
    segmT = duration/num
    waveSegs = [wave.segment(i,i+segmT) for i in range(0,duration,segmT)] 
    return waveSegs

def computeDCTSegs(waveSegs):
    
    dctSegs = []
    for i in waveSegs:
        floats = [float(y) for y in i.ys]
        amps = scipy.fftpack.dct(floats, type=2)
        dct = thinkdsp.Dct(amps, i.framerate)
        dctSegs.append(dct)
    return dctSegs
    

def cullFreqs(dctSegs, amp=1000):
    culled = []
    counts = []
    for dctVal in dctSegs:
        dctlist = dctVal
        count = 0
        for i,x  in enumerate(dctVal.amps):
            if x<amp:
                if i<dctVal.amps.size():
                    dctlist.amps = numpy.delete(dctVal.amps, i, None)
                    count+=1
        counts.append(count)
        culled.append(dctlist)
    return culled, counts

def invDCT(culled):
    wavs = [i.make_wave() for i in culled]
    
    for i,wav in enumerate(wavs):
        if i == 0:
            wavi = wav
        else:
            wavi = wavi|wav
    return wavi

waveSegs = breakWave(wave)
dctSegs = computeDCTSegs(waveSegs)
cull,count = cullFreqs(dctSegs)
wave2 = invDCT(cull)
wave2.make_audio()
print count
```
