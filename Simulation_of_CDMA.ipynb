```python
import numpy as np
import pyaudio
from array import array
import matplotlib.pyplot as mplib

# This function converts a string into a numpy array of bits
# note that it is assumed that each character is 7 bits long here
def string2NPArray(s):
    bits = np.array([])
    for a in bytearray(s, 'ascii'):
        for b in range(0,7):
            bits = np.append(bits,float((a>>(7-b-1))&1))
    return bits

# This function converts a numpy array of bits to a string
# note that it is assumed that each character is 7 bits long here
def NPbits2String(bits):
    S = ""
    for a in np.arange(0, np.ceil(len(bits)/7)):
        tmp = 0
        for k in np.arange(0,7):
            b = bits[a*7+k]
            tmp = tmp + (2**(6-k))*b
        S = S + chr(int(tmp))
    return S


# this function plays samples through the speaker
# at sample rate of rate
def play_samples(x, rate=8000, chunk_size = 1000):
    p = pyaudio.PyAudio()
    stream = p.open(format= pyaudio.paInt16, channels=1, rate=rate,
        input=False, output=True,
        frames_per_buffer=len(x))
    stream.write(x.astype(np.int16).tostring())
    stream.stop_stream()
    stream.close()
    p.terminate()
    

# this is a utility function that  finds the start and  end 
# of transmission in the numpy array of samples xf
# The function looks for the first instance where the entries of xf
# go above threshold and returns the index into xf where this happens
# in start_idx
# The function looks for the last instance where the entries of xf
# are above threshold and returns the index into xf where this happens
# in end_idx
# 
# You will probably have to do some trial and error to get the threshold right
# one possibility is to se the threshold equal to some factor of the maximum value
# in the input signal,  e.g. 0.3 * maximum value in xf
#
def find_start_and_end(xf, threshold = 2000): 
    import numpy as np    
    start_idx = -1
 
    for k in range(0, len(xf)):
        if(np.abs(xf[k])) > threshold:
            start_idx = k
            break

    if(start_idx  < 0):
        print "Unable to detect start of transmission"
        return -1
    
    end_idx = -1
    
    for k in range(0, len(xf)):
        if(np.abs(xf[len(xf)-k-1])) > threshold:
            end_idx = len(xf)-k-1
            break

    if(end_idx < 0):
        print "Unable to detect end of transmission"
        return -1

    return start_idx, end_idx
```

We can generate codes by taking the rows or columns of the hadamard matrix, which we can generate using the follow recurssive equation:
$$H(2^k) = \begin{bmatrix}
H(2^{k-1}) &  H(2^{k-1})\\
H(2^{k-1})  & -H(2^{k-1})\end{bmatrix} $$

```python
def Hadamard(x, sign=1):
#     first implementation of Hadamard matrix
    """Generates Hadamard matrix recursively, x is the length of the matrix, sign is the sign"""
    if x == 1 and sign == 1:
        return [1,1,1,-1]
    if x == 1 and sign == -1:
        return [-1, -1, -1, 1]
    if sign == 1:
        return [Hadamard(x-1,1), Hadamard(x-1,1), Hadamard(x-1,1), Hadamard(x-1,-1)]
    if sign == -1:
        return [Hadamard(x-1,-1), Hadamard(x-1,-1), Hadamard(x-1,-1), Hadamard(x-1,1)]

def walsh(x, sign=1):
    #second pass at code generation
    """Generates a array of Walsh codes, x is the length of the code, sign is the sign"""
    if x == 1 and sign == 1:
        return [[1,1],[1,-1]]
    if x == 1 and sign == -1:
        return [[-1, -1],[-1, 1]]
    a = walsh(x-1,1)
    b = walsh(x-1,-1)
    if sign == 1:
        out = []
        for i in range(len(a)*2):
            if i<len(a):
                out.append(a[i]+a[i])
            else:
                out.append(a[i-len(a)]+b[i-len(a)])
        return out
    if sign == -1:
        out = []
        for i in range(len(b)*2):
            if i<len(b):
                out.append(b[i]+b[i])
            else:
                out.append(b[i-len(b)]+a[i-len(b)])
        return out

def dotProduct(a,b):
    """Finds the dot product of two lists, list must contain only numbers and be of the same length. a and b must both be lists"""
    #I was lazy and decided to implement my own dot product function
    if len(a)!=len(b):
        return False
    out = 0
    for i in range(len(a)):
        out += a[i]*b[i]
    return out
```

```python
import random

def generateCodeAsync(N):
    """generate the code for asynchronous CDMA"""
    output = []
    for i in range(N):
        num = random.random()
        if num<.5:
            output.append(-1)
        else: 
            output.append(1)
    return output

def outer1D(a, b):
    """computes the outer product in a 1D list"""
    output = []
    for i in a:
        for j in b:
            output.append(i*j)
    return output

def fakeTransmit(a,b):
    """adds the signals as if they were in the air at the same time"""
    output = []
    for i in range(len(a)):
        output.append(a[i]+b[i])
    return output
```
